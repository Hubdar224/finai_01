```python
import seaborn as sns
import matplotlib.pyplot as plt 

boat = sns.load_dataset("titanic")
sns.boxplot(x="survived", 
            y="age", showmeans=True, 
            meanprops= {"marker":"*", "markersize":"12", "markeredgecolor":"red"}, 
            data=boat)
plt.xlabel("how many", size=12),
plt.ylabel("Age(years)", size=12),
plt.title("kitny bachy or kitny doby", size=12, weight='bold')
plt.show()
```
